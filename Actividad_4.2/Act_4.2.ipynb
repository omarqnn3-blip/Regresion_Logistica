```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('projectos_forvia.csv')
```

```python
df = df.drop(df.index[245]).reset_index(drop=True)
```

```python
df.info()
df.isnull().sum()
```

```python
df["Percent complete"] = df["Percent complete"].astype(str).str.replace("%", "", regex=False)
df["Percent complete"] = pd.to_numeric(df["Percent complete"], errors="coerce")
```

```python
df = df.drop(columns=["Actual end date", "Closed", "Domain Path", "Actual Go Live date"])
df.info()
```

#### **Procesamos outliers y nulos** 

```python
df["Planned start date 2"] = pd.to_datetime(df["Planned start date"], dayfirst=True, errors='coerce')
```

```python
sum = int(df["Planned start date 2"].isnull().sum())
sum
```

```python
df["Planned start date 3"] = df["Planned start date"].str.extract(r'(\d{1,2}/\d{1,2}/\d{4})')
```

```python
sum = int(df["Planned start date 3"].isnull().sum())
sum
```

```python
df[df["Planned start date 2"].isnull()][["Planned start date", "Planned start date 2", "Planned start date 3"]]
```

```python
df['Planned start date'] = pd.to_datetime(df['Planned start date 3'], format='%d/%m/%Y', dayfirst=True, errors='coerce')
```

```python
df[df["Planned start date 2"].isnull()][["Planned start date", "Planned start date 2", "Planned start date 3"]]
```

```python
df.drop(columns=["Planned start date 2", "Planned start date 3"], inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
df["Project target phase"] =df["Project target phase"].fillna("NA") 
df["Last WAR"] =df["Last WAR"].fillna("NA")
df["Actual start date"] =df["Actual start date"].fillna("NA")
```

```python
df["Project Health"] =df["Project Health"].fillna('Red')
```

```python
df.isnull().sum()
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['Project Type', 'Geographical scope', 'Project manager', 'State', 'Project size', 'Project organization', 'BG', 'Project Health', 'On-hold']
for col in cols:
    df[col] = le.fit_transform(df[col])
    df[col] = df[col] + 1  # Incrementar en 1 para que la mayor frecuencia sea 1
df_num = df.select_dtypes(include=[np.number])
df_num.info()
```

### **Regresion Logistica**

### On hold

```python
df_num.columns
```

```python
#Declaramos las variables dependientes e independientes para la regresión Logística
Vars_Indep= df_num[['Project Type', 'Geographical scope', 'Project size']]
Var_Dep= df['On-hold']
```

```python
#Redefinimos las variables 
X= Vars_Indep
y= Var_Dep
```

```python
from sklearn.model_selection import train_test_split

#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =None)
```

```python
from sklearn.preprocessing import StandardScaler

#Se escalan todos los datos
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las variables “X” tanto de entrenamiento como de prueba, utilizaremos fit_transform
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Reponderación de las clases
algoritmo_Pond = LogisticRegression(class_weight='balanced')
algoritmo_Pond.fit(X_train, y_train)
```

```python
#Realizamos una predicción
y_pred_pond = algoritmo_Pond.predict(X_test) 
y_pred_pond
```

```python
#Verifico la matriz de Confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred_pond)
print('Matriz de Confusión:')
print(matriz)
```

```python
#con small
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred_pond, average="binary", pos_label='SMALL')
print('Precisión del modelo:')
print(precision)
```

```python
#con 2
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred_pond, average="binary", pos_label='LARGE/MEDIUM')
print('Precisión del modelo:')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred_pond)
print('Exactitud del modelo:')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred_pond, average="binary", pos_label='SMALL')
print('Sensibilidad del modelo:')
print(sensibilidad)
```

```python
#Calculo la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred_pond, average="binary", pos_label='LARGE/MEDIUM')
print('Sensibilidad del modelo:')
print(sensibilidad)
```

```python
# el puntaje F1 que es una combinación entre la precisión y la sensibilidad, para esto importamos f1_score.
#Calculo el Puntaje F1 del modelo
from sklearn.metrics import f1_score

puntajef1 = f1_score(y_test, y_pred_pond, average="binary", pos_label=1)
print('Puntaje F1 del modelo:')
print(puntajef1)
```

```python
algoritmo = LogisticRegression()
algoritmo.fit(X_train, y_train)
```
